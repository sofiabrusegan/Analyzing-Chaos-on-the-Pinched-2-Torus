### U(r)

```python
def U(r, pphi, c):
    return ((pphi**2)/(2*r**2)) - 0.5*(r**2) + 0.25*(r**4) + c*pphi

rs = np.linspace(-2, 2, 1000)
pphis = [0, 0.1, 0.3, 0.5, 0.7]

plt.figure()

plt.plot(rs, [0]*len(rs), color = 'silver')
plt.plot([0]*len(rs), rs, color = 'silver')

for pphi in pphis:
    urs = U(rs, pphi, 0)
    plt.plot(rs, urs, label = rf"$p_\phi = {pphi:.01f}$")

plt.xlabel("$r$")
plt.ylabel("$U(r)$")
plt.legend()
plt.xlim(-2, 2)
plt.ylim(-0.5, 2)
plt.savefig("Effective potential c0.pdf", format="pdf", bbox_inches = 'tight')
```

### Poincare plot and conservation of H

```python
# -------------------------- 2-torus plot --------------------------
def r(t):
    return math.sqrt(2) / np.cosh(t)

def pr(t):
    return (-math.sqrt(2) * np.tanh(t))/np.cosh(t)

tvals = np.linspace(-100, 100, 1000)
rvals_torus = []
prvals_torus = []

for t in tvals:
    rvals_torus.append(r(t))    
    prvals_torus.append(pr(t))

plt.figure()
plt.plot(rvals_torus, prvals_torus, "--", color = 'black',label = "Pinched 2-torus")

# -------------------------- Plot values --------------------------
a = 0
b = 1000
num_points = 100
c_val = 1
epsilon = 0.0

# -------------------------- Poincare algorithm --------------------------

num_t = 1000
rs = np.linspace(1.5, 1, num_t)

# Stores the r values to be used as initial conditions
rvals = []
prvals = []

# Choose set of values depending on c_val
#pphis = [0.1, 0.3, 0.5, 0.7]
pphis = [0, 0.1, 0.2, 0.21]
colors = ['b', 'r', 'c', 'y']
assert len(pphis) == len(colors)

for pphi in pphis:
    for j, r in enumerate(rs):
        if pr_squared(r, 0, pphi, c = c_val, eps = epsilon) >= 0:
            new_rs = np.linspace(rs[j-1], rs[j], num_t)
            break
    for k in range(len(new_rs)):
        r = new_rs[k]
        pr_square = pr_squared(r, 0, pphi, c = c_val, eps = epsilon)
        if pr_square >= 0:
            final_r = r
            final_pr = math.sqrt(pr_square)
            break
    rvals.append(final_r)
    prvals.append(final_pr)

rvals = np.array(rvals)

#--------------------------------------------------------------

# We would like to store all of the values plotted to see how accurate they are
# This is done by seeing how well the energy H = 0 is conserved

# We create an array (4*len(pphis)) x (2*num_points)
# We use 2*num_points in case more points need to be stored, which is unlikely yet possible

store_arr = np.zeros((4*len(pphis), 2*num_points))

#--------------------------------------------------------------
for i, r in enumerate(rvals):
    print(f"Starting {i+1}...")
    z0 = np.array([r, 0, prvals[i], pphis[i]])
    x, y, pphis1 = henon_alg_polar(polar_dynamics, henon_f, a, b, z0, tol=1e-8, N = num_points, max_it = 5, max_step_user=1e-2, c = c_val, eps = epsilon)   

    poincare_r = []
    poincare_pr = []
    poincare_pphis = []

    for k in range(len(x)):
        if abs(H_polar(np.array([x[k], 0, y[k], pphis1[k]]), c = c_val, eps = epsilon)) < 1e-2:
            poincare_r.append(x[k])
            poincare_pr.append(y[k])
            poincare_pphis.append(pphis1[k])


    # We store these values in the array to later check their accuracy
    store_arr[4*i, 0:len(poincare_r)] = poincare_r
    store_arr[4*i+2, 0:len(poincare_pr)] = poincare_pr
    store_arr[4*i+3, 0:len(poincare_pphis)] = poincare_pphis
    
    plt.scatter(x, y, s=6, label=rf"$p_\phi$ = {pphis[i]:.02f}", color=colors[i])

plt.xlim(-0.1,1.5)
plt.ylim(-1, 1)
plt.legend()
plt.xlabel("$r$")
plt.ylabel("$p_r$")
plt.savefig(f"Poincare map c{c_val} eps{epsilon}.pdf", format="pdf", bbox_inches = 'tight')
#plt.show()

# -------------------------- Conservation of H --------------------------
plt.figure()

for i in range(len(pphis)):
    idx = 0
    H_vals = []
    while store_arr[4*i, idx] != 0:
        z = store_arr[4*i:4*i + 4, idx]
        H_vals.append(H_polar(z, c = c_val, eps = epsilon))
        idx += 1
    plt.plot(range(len(H_vals)), H_vals, label = rf"$p_\phi$ = {pphis[i]:.02f}", color = colors[i])

plt.legend()
plt.xlabel("$n$th point plotted")
plt.ylabel(r"$\tilde {H}^\varepsilon$")
#plt.show()
plt.savefig(f"Conservation of H along Poincare c{c_val} eps{epsilon}.pdf", format="pdf", bbox_inches = 'tight')
```

### Cartesian integration and conservation of H

```python
# -------------------------- 2-torus plot --------------------------
def r(t):
    return math.sqrt(2) / np.cosh(t)

def pr(t):
    return (-math.sqrt(2) * np.tanh(t))/np.cosh(t)

tvals = np.linspace(-100, 100, 1000)
rvals_torus = []
prvals_torus = []

for t in tvals:
    rvals_torus.append(r(t))    
    prvals_torus.append(pr(t))

plt.figure()
plt.plot(rvals_torus, prvals_torus, "--", color = 'black',label = "Pinched 2-torus")

# -------------------------- Plot values --------------------------
a = 0
b = 1000
num_points = 200
c_val = 1
epsilon = 0.01

# -------------------------- Finding initial conditions algorithm --------------------------

num_t = 1000
rs = np.linspace(1.5, 0.01, num_t)

# Stores the r values to be used as initial conditions
rvals = []
prvals = []

# Choose set of values depending on c_val
#pphis = [0.1, 0.3, 0.5, 0.7]
pphis = [0, 0.1, 0.2, 0.21]
colors = ['b', 'r', 'c', 'y']
#colors = ['b', 'r', 'c', 'y']
assert len(pphis) == len(colors)

for pphi in pphis:
    for j, r in enumerate(rs):
        if pr_squared(r, 0, pphi, c = c_val, eps = epsilon) >= 0:
            new_rs = np.linspace(rs[j-1], rs[j], num_t)
            break
    for k in range(len(new_rs)):
        r = new_rs[k]
        pr_square = pr_squared(r, 0, pphi, c = c_val, eps = epsilon)
        if pr_square >= 0:
            final_r = r
            final_pr = math.sqrt(pr_square)
            break
    rvals.append(final_r)
    prvals.append(final_pr)

rvals = np.array(rvals)

#--------------------------------------------------------------

# We would like to store all of the values plotted to see how accurate they are
# This is done by seeing how well the energy H = 0 is conserved

store_arr = np.zeros((4*len(pphis), 2*num_points))
# A point is stored along [4i : 4(i+1), idx]

# -------------------------- Poincare algorithm --------------------------
for j, r in enumerate(rvals):
    print(f"Starting {j+1}...")
    z0 = [r, 0, prvals[j], pphis[j]] #in polar
    z0cart = polar_to_cart(z0) # in cartesian
    x,y, px, py = henon_alg_cart(cart_dynamics, henon_f_cart, a, b, z0cart, 
                                 tol = 1e-8, N = num_points, max_it = 3, max_step_user = 1e-2,
                                 c = c_val, eps = epsilon)

    r_poincare = []
    pr_poincare = []

    for k in range(len(x)):
        cartPoincare = np.array([x[k], y[k], px[k], py[k]]) # poincare value in cartesian
        polarPoincare = cart_to_polar(cartPoincare) # poincare value in polar
        polarPoincare[1] = 0 # since assumption is phi = 0, set it equal to make evaluation accurate
        
        # For the Poincare plot
        r = polarPoincare[0]
        pr = polarPoincare[2]
        r_poincare.append(r) # for r vs pr poincare plot
        pr_poincare.append(pr)

        # Stored to see how well H is conserved
        store_arr[4*j : 4*j+4, k] = polarPoincare # store in array to later plot
    
    plt.scatter(r_poincare, pr_poincare, s = 6, label = rf"Initial $p_\phi$ = {pphis[j]:.02f}", color = colors[j])

plt.legend()
plt.xlim(-0.1,1.5)
plt.ylim(-1, 1)
plt.xlabel("$r$")
plt.ylabel("$p_r$")
plt.savefig(f"Cartesian Poincare c{c_val} eps{epsilon}.pdf", format="pdf", bbox_inches = 'tight')
#plt.show()

# -------------------------- Conservation of H plot --------------------------

plt.figure()

for l in range(len(pphis)):
    idx = 0
    H_vals = []
    while store_arr[4*l, idx] != 0:
        r, _, pr, pphi = store_arr[4*l : 4*l +4, idx]
        H_vals.append(H_polar(np.array([r, 0, pr, pphi]), c = c_val, eps = epsilon))
        idx += 1
    plt.plot(range(len(H_vals)), H_vals, label = rf"Initial $p_\phi$ = {pphis[l]:.02f}", color = colors[l])

plt.legend()
plt.xlabel("$n$th point plotted")
plt.ylabel(r"$\tilde {H}^\varepsilon$")
plt.savefig(rf"Cartesian conservation of H c{c_val} eps{epsilon}.pdf", format="pdf", bbox_inches = 'tight')
#plt.show()
```
